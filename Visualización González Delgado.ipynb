```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
import os
drive.mount('/content/gdrive')
%cd '/content/gdrive/MyDrive/Proyecto DSCH/suicidios'
suicidios = pd.read_csv('master.csv')
```

###Graficamos con Matplotlib
- Cantidad total de suicidios según sexo
- Cantidad total de suicidios por año
- Cantidad total de suicidios por país

```python

suicidios = suicidios.reset_index(drop=True)
fig, ax = plt.subplots(figsize=(12, 12))
agg_suicidios = suicidios.groupby(['sex']).sum()
agg_suicidios['suicides_no'].plot.bar(stacked=True)
plt.xlabel('Sexo')
plt.ylabel('Cantidad de suicidios')
plt.title('Cantidad de suicidios según sexo')
```

```python
suicidios = suicidios.reset_index(drop=True)
fig, ax = plt.subplots(figsize=(20, 8))
agg_suicidios = suicidios.groupby(['year']).sum()
agg_suicidios['suicides_no'].plot.line()
plt.xlabel('Año')
plt.ylabel('Cantidad de suicidios')
plt.title('Cantidad de suicidios por año')
plt.xticks(rotation='vertical')
plt.xlim((1985, 2015))
```

```python
suicidios = suicidios.reset_index(drop=True)
fig, ax = plt.subplots(figsize=(30, 8))
agg_suicidios = suicidios.groupby(['country']).sum()
agg_suicidios['suicides_no'].plot.bar()
plt.xlabel('País')
plt.ylabel('Cantidad de suicidios')
plt.title('Cantidad de suicidios por país')
```

### Graficamos con Seaborn

- Cantidad total de suicidios por año
- Cantidad de  total suicidios por país según sexo
- Cantidad total de suicidios por rango de edades según sexo

```python

f = suicidios.groupby('year').sum()
fig, ax = plt.subplots(figsize=(30, 8))
sns.lineplot(data=f, x="year", y='suicides_no')
plt.xticks(rotation=45)
plt.xlabel('Año')
plt.ylabel('Cantidad de suicidios')
plt.title('Cantidad de suicidios por año')
plt.xlim((1985, 2015))
```

```python
fig, ax = plt.subplots(figsize=(30, 8))
sns.barplot(data=suicidios, x="country", y="suicides_no", hue='sex', ci=None)
plt.xticks(rotation = 90)
plt.xlabel('Pais')
plt.ylabel('Número de Suicidios')
plt.title('Cantidad de suicidios por país según sexo')
```

```python
fig, ax = plt.subplots(figsize=(30, 8))
sns.barplot(data=suicidios, x="age", y='suicides_no', hue='sex', order=['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years'],ci=None)
plt.xticks(rotation=45)
plt.xlabel('Año')
plt.ylabel('Cantidad de suicidios')
plt.title('Cantidad total de suicidios por rango de edades según sexo')
```
