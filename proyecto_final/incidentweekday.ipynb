```python
#importamos librerías necesarias y generamos dataframe a partir de los datos
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import os
import requests
import json
import plotly.express as px
from plotly.subplots import make_subplots

#generamos dataframe
gun_weather = pd.read_csv('C:\coderhouse\coderhouse\proyecto_final\gun_violence_full.csv')
```

```python
gun_state_ill_day = gun_state_ill.groupby('day_of_the_week').sum()
```

```python
fig, ((ax1, ax2, ax3)) = plt.subplots(1, 3, figsize=(24, 10))
ax1.plot(gun_state_ill_day['incident_amount'], color="orange")
ax2.plot(gun_state_ill_day['n_injured'],color='red')
ax3.plot(gun_state_ill_day['n_killed'], color="blue")

fig.suptitle("Estadísticas Diarias", size=16)
labels = [0,'Lunes', 'Martes', 'Miércoles','Jueves', 'Viernes', 'Sábado', 'Domingo',]
ax1.set_title('Incidentes cada día')
ax2.set_title('Heridos cada día')
ax3.set_title('Muertos cada día')
ax1.set_xlabel('Día de la semana')
ax2.set_xlabel('Día de la semana')
ax3.set_xlabel('Día de la semana')
ax1.set_ylabel('Incidentes')
ax2.set_ylabel('Heridos')
ax3.set_ylabel('Muertos')
ax1.set_xticklabels(labels, rotation='vertical')
ax2.set_xticklabels(labels, rotation='vertical')
ax3.set_xticklabels(labels, rotation='vertical')

```

```python
fig, ((ax1, ax2, ax3)) = plt.subplots(1, 3, figsize=(24, 10))
ax1.plot(gun_state_ill_month['incident_amount'], color="green")
ax2.plot(gun_state_ill_month['n_injured'],color='black')
ax3.plot(gun_state_ill_month['n_killed'], color="red")

fig.suptitle("Estadísticas Mensuales", size=16)
labels2 = [0, 'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Setiembre', 'Octubre', 'Noviembre', 'Diciembre']
ax1.set_title('Incidentes cada mes')
ax2.set_title('Heridos cada mes')
ax3.set_title('Muertos cada mes')
ax1.set_xlabel('Mes del año')
ax2.set_xlabel('Mes del año')
ax3.set_xlabel('Mes del año')
ax1.set_ylabel('Incidentes')
ax2.set_ylabel('Heridos')
ax3.set_ylabel('Muertos')
ax1.locator_params(axis= 'x',nbins=24)
ax2.locator_params(axis= 'x',nbins=24)
ax3.locator_params(axis= 'x',nbins=24)
ax1.set_xticklabels(labels2, rotation='vertical')
ax2.set_xticklabels(labels2, rotation='vertical')
ax3.set_xticklabels(labels2, rotation='vertical')
```
